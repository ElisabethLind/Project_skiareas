#### Wie groß sind Schigebiete? 
##### Pistenkilometer 
- arithmetisches Mittel und Median
- größter Wert, kleinster Wert 
- mach mir ein paar neue Gruppen 
    - 0 - 10
    - 10.1 - 30
    - 30.1 - 100
    - ab 100.1 


```python
# wie groß ist das arithmetische Mittel? 
df["Pistenkilometer"].mean()
```

```python
#  wie groß ist der Median?
df["Pistenkilometer"].median()
```

```python
# Was sind die größten Schigebiete?
df_biggest = df.sort_values(by="Pistenkilometer", ascending=False)
df_biggest.head()
```

```python
# und was sind die kleinsten?
df_biggest = df.sort_values(by="Pistenkilometer", ascending=False)
df_biggest.tail()
```

```python
# ich definiere mir eine function um die Pistenkilometer in vier Gruppen zu unterteilen 
def count_pistenkilometer_ranges(df):
    # ich definiere die Ranges und summiere sie 
    ranges = {
        "0 to 10": ((df['Pistenkilometer'] >= 0) & (df['Pistenkilometer'] <= 10)).sum(),
        "10.1 to 30": ((df['Pistenkilometer'] > 10) & (df['Pistenkilometer'] <= 30)).sum(),
        "30.1 to 100": ((df['Pistenkilometer'] > 30) & (df['Pistenkilometer'] <= 100)).sum(),
        "more than 100.1": (df['Pistenkilometer'] > 100).sum()
    }
    # und geb sie aus 
    return ranges


# ich führe die Funktion aus 
count_pistenkilometer_ranges(df)
```

```python
# ich mach mir daraus ein dataframe 
# zuerst definiere ich mein Ergebnis als Variable 
table1 = count_pistenkilometer_ranges(df)
table1

# dann mach ich mir ein dataframe, das ich seehohe nenne 
# from_dict (= weil das ein dictionary ist)
# table1 = nimm die table 1
# orient='index' heißt die heys des dict sind die rows
# column ist Pistenkilometer 
pistenkilometer_df = pd.DataFrame.from_dict(table1, orient='index', columns=['Pistenkilometer'])

pistenkilometer_df
```

```python
#ich mach mir ein excel draus, das ich dann in den datawrapper geben kann 
pistenkilometer_df.to_excel("Pistenkilometer.xlsx")
```

#### Wie viele Schigebiete gibt es in den Bundesländern?

```python
# Wie viele Schigebiete gibt es pro Bundesland?
bundesland_counts = df['Bundesland'].value_counts()
bundesland_counts
```

#### Wie viele davon liegen über 2.000m?

```python
# jetzt mach ich mir eine Variable über 2000
above_2000 = df[df['Seehöhe (in m)'] > 2000]

# und jetzt zähle ich alle elemente über 2000 
bundesland_counts_above_2000 = above_2000['Bundesland'].value_counts()
bundesland_counts_above_2000
```

```python
# ich hätte jezt gern beide ergebnisse in einer neuen Tabelle zusammengefügt 
# 2 columns = Schigebiete total, Schigebiete über 2000 m 
# rows = Bundesländer 

combined_df = pd.DataFrame({
    'Schigebiete total': bundesland_counts,
    'Schigebiete über 2000': bundesland_counts_above_2000
})

# ich ersetzte NaN drch 0, um es besser lesen zu können 
combined_df.fillna(0, inplace=True)

# ich mach mir aus beiden columns integers, um keine Kommastellen zu haben 
combined_df['Schigebiete total'] = combined_df['Schigebiete total'].astype(int)
combined_df['Schigebiete über 2000'] = combined_df['Schigebiete über 2000'].astype(int)

combined_df
```

```python
#ich mach mir ein excel draus, das ich dann in den datawrapper geben kann 
combined_df.to_excel("Bundeslander.xlsx")
```
