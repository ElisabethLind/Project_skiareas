### Skigebiete Österreichs 
- Quelle: https://www.bergfex.at/
- 10 Seiten, 395 Skigebiete  (40 Hits/Seite)

```python
# ich importiere alles, was ich brauchen werde 
import requests
import pandas as pd 
import re 
from bs4 import BeautifulSoup
import numpy as np
```

```python
# ich nehme mir die erste Seite und schaue sie mir an 
response = requests.get("https://www.bergfex.at/oesterreich/?page=1")
# gib mir den status code 
response
```

```python
ski_data = response.text
print(ski_data)
```

```python
# ich mach mir aus der Seite 1 ein soup-element
soup = BeautifulSoup(ski_data)
type(soup)
```

```python
# ich schau mir das element an 
soup
```

```python
# ich schau mir mal für die erste Seite die Namen der Schigebiete an 
# Untersuchen auf der HP: class = @3xl:tw-overflow-hidden @3xl:tw-grow-0 tw-mr-1 @3xl:tw-mr-0
name = soup.find_all('span', class_="@3xl:tw-overflow-hidden @3xl:tw-grow-0 tw-mr-1 @3xl:tw-mr-0")
len(name)
```

```python
# Höhenmeter: w-hidden @3xl:tw-block tw-ml-1 tw-text-s tw-whitespace-nowrap @3xl:tw-text-m @3xl:tw-w-13 @3xl:tw-ml-0 @3xl:tw-shrink-0
height = soup.find_all('div', class_="tw-hidden @3xl:tw-block tw-ml-1 tw-text-s tw-whitespace-nowrap @3xl:tw-text-m @3xl:tw-w-13 @3xl:tw-ml-0 @3xl:tw-shrink-0")
len(height)
```

```python
# Pistenkilometer: tw-flex tw-items-center tw-gap-2 @3xl:tw-gap-0 tw-mt-2 @3xl:tw-mt-0
km = soup.find_all('div', class_="tw-flex tw-items-center tw-gap-2 @3xl:tw-gap-0 tw-mt-2 @3xl:tw-mt-0")
len(km)
```

```python
# Lifte: tw-hidden md:tw-inline
# da ist das erste Element zum wegschmeißen 
lifte = soup.find_all('span', class_="tw-hidden md:tw-inline")
len(lifte)
```

```python
# warum habe ich da 41 Ergbenisse? 
# erstes element schaut wie aus? 
lifte[0]
```

```python
# zweites element? 
lifte[1]
```

```python
# das erste element brauchen wir also nicht 
# Wichtig: ich hab das für die Seite 2 auch durchgespielt - und man muss das bei allen Seiten machen! 
lift = lifte[1:41]
lift
```

```python
# ich mache mir eine Liste, in die ich alles zusammenlege - also alle elemente, die ich vorher rausgescraped habe
# leere Liste skiing, die wir dann befüllen 
skiing = []

# for loop
# zip() ist eine function in python  mit der ich variablen definieren kann, die mir zusammengehängt werden in einer Liste
# also hier definiere ich meine elemente und gebe an wie sie heißen nach dem zip() - das sind meine variablen, die ich vorher definiert habe  
for elm1,elm2,elm3,elm4 in zip(name, height, km, lift):
    # ich ordne die elemente den Variablen zu 
    ski = {
        "name": elm1,
        "height": elm2,
        "km": elm3,
        "lift": elm4
    }
    # ich sag häng mir bitte alles zusammen 
    skiing.append(ski)

# und gib mir die fertige Liste aus 
skiing
```

### Step 2: dataframe machen

```python
# ich mach mir ein datframe aus der Tabelle
df = pd.DataFrame(skiing)
df.head()
```

### Step 3: Code für alle 10 Seiten

```python
# Step 1: Vorbereitungen 
# ich mach mir dazu ein base_url (ohne Seitennummer)
base_url = "https://www.bergfex.at/oesterreich/?page="

# und ich mach mir eine neue, leere Liste 
skiing = []

# Step 2: jetzt loope ich durch jede Seite (also page in range 1 -11: wichtig, die letzte page wird nicht mehr gemacht - deshalb 11!)
# alles andere ist im loop 
for page in range(1, 11):
    # Step 3: 
    # ich definiere meine url - das ist die basis_url (siehe oben und die page)
    url = f"{base_url}{page}"
    # ich sag zu requests gib mir die url
    response = requests.get(url)
    # ich sag mach mir daraus also aus response ein BeautifulSoup-element 
    # .content sagt einfach das ich gerne den Inhalt der response hätte 
    # html.parser (= python-ding) hier wicjtig, damit man mit dem html-code arbeiten kann
    soup = BeautifulSoup(response.content, 'html.parser')

    # Step 4: hier kommen jetzt meine Variablen zum Einsatz, die ich vorher definiert habe 
    names = soup.find_all('span', class_="@3xl:tw-overflow-hidden @3xl:tw-grow-0 tw-mr-1 @3xl:tw-mr-0")
    heights = soup.find_all('div', class_="tw-hidden @3xl:tw-block tw-ml-1 tw-text-s tw-whitespace-nowrap @3xl:tw-text-m @3xl:tw-w-13 @3xl:tw-ml-0 @3xl:tw-shrink-0")
    kms = soup.find_all('div', class_="tw-flex tw-items-center tw-gap-2 @3xl:tw-gap-0 tw-mt-2 @3xl:tw-mt-0")
    lifte = soup.find_all('span', class_="tw-hidden md:tw-inline")

    # bei lifts nimm bitte nur element 1 bis 41 (nicht element 0)
    lifts = lifte[1:41]

    # Step 5: jetzt hänge ich meine elemente zusammen 
    for elm1, elm2, elm3, elm4 in zip(names, heights, kms, lifts):
        ski = {
            "name": elm1.get_text(strip=True),
            "height": elm2.get_text(strip=True),
            "km": elm3.get_text(strip=True),
            "lift": elm4.get_text(strip=True)
        }
        # und leg mir das alles in meine liste skiing 
        skiing.append(ski)

# und dann hätt ich gern ein Dataframe 
df = pd.DataFrame(skiing)
df

```

### Step 4: Cleaning

```python
# wir säubern jetzt noch die column lift
# wir ersetzen das Wort Lifte durch nichts!
df['lift'] = df['lift'].str.replace('Lifte', '') 

# wir ersetzen \n durch nichts!
df['lift'] = df['lift'].str.replace('\n', '')  

# ich entferne in der column km die km 
df["km"] = df["km"].str.replace("km", "")

# ich entferne in der column Seehöhe die m 
df["height"] = df["height"].str.replace("m", "")
df["height"] = df["height"].str.replace(".", "")

# wir konvertieren alles in integer in dieser column (pd.to_numeric(df['lift']) und dort, wo nichts steht bzw. wo Ergebnisse stehen, die man nicht umwandeln kann, sagen wir schreib bitte NaN (Not a Number) (errors='coerce')
df['lift'] = pd.to_numeric(df['lift'], errors='coerce')

# schreib mir statt NaN 0
df['lift'].fillna(0, inplace=True)

# mach mir aus de ganzen column einen integer 
df['lift'] = df['lift'].astype(int)

# bei km und height habe ich zT statt o ein - 
# das entferne ich zuvor noch 
df['km'] = df['km'].replace('-', '0').astype(float)

df['height'] = df['height'].replace('-', '0').astype(int)


print(df)
```
